```python
import numpy as np
import pandas as pd
from scipy.optimize import fsolve
```

```python
D = pd.read_excel('Extra_Credit_Term_Structure.xlsx', header=None).values.reshape(-1)
D = {i:D[int(i*2-1)] for i in np.arange(0.5,15.5,0.5)}
vol = pd.read_excel('Extra_Credit_Voldat.xlsx', header=None).values.reshape(-1)
```

```python
R = []
tau = np.sqrt(0.5)
```

```python
def obj1(x): #for fitting D(0.5)
    return 2*((1/D[0.5])-1)-x
R.append(list(fsolve(obj1,0.1)))
R
```

```python
def objective(r_star, R, end_nodes, vol_index, D_index, tau):
    time_steps = end_nodes
    end_prices = []
    previous_prices = []
    for i in range(time_steps):
        temp = []
        for ii in range(end_nodes - i):
            if(i == 0):
                if(ii==0):
                    end_prices.append(1/(1+(r_star/2)))
                else:
                    calc = r_star*np.exp(-ii*2*vol[vol_index]*tau)
                    end_prices.append(1/(1+(calc/2)))
            elif(i == 1):
                temp.append((0.5*end_prices[ii] + 0.5*end_prices[ii+1])/(1+R[time_steps-2][ii]/2))
            else:
                temp.append((0.5*previous_prices[i-2][ii] + 0.5*previous_prices[i-2][ii+1])/(1+R[time_steps-i-1][ii]/2))
        if(i != 0):
            previous_prices.append(temp)
    return previous_prices[-1][0] - D[D_index]

def rates(objective, intial_guess, R, end_nodes, vol_index, D_index, tau ):
    r_star = fsolve(objective, intial_guess, args=(R, end_nodes, vol_index, D_index, tau))[0]
    temp = []
    for i in range(end_nodes):
        if(i==0):
            temp.append(r_star)
        else:
            temp.append(r_star*np.exp(-i*2*vol[vol_index]*tau))
    return temp
```

```python
for i in range(29):
    R.append(rates(objective, intial_guess=0.1, R=R, end_nodes=i+2, vol_index=i, D_index=(1+i/2), tau=tau))
```
